1 -  Function to plot and save distribution of each column

'/content/drive/MyDrive/IE/CORPORATE_PROJECT_AIRBUS/02_Notebooks/pics

```python
import os
import matplotlib.pyplot as plt

def plot_and_save_distribution(dataframe, folder_path='pics'):
    os.makedirs(folder_path, exist_ok=True)

    for column in dataframe.columns:
        plt.figure(figsize=(10, 6))
        dataframe[column].plot(kind='hist', bins=50, title=f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.savefig(os.path.join(folder_path, f'{column}_distribution.jpeg'))
        plt.close()

```

```python
plot_and_save_distribution(df, '/content/drive/MyDrive/IE/CORPORATE_PROJECT_AIRBUS/02_Notebooks/pics')
```
