```python
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.max_rows', None)  # Display all rows


data1 = {}
data1['Features'] = list(msn02.columns)
data1['msn02'] = list(msn02.isnull().sum())
data1['msn02_%'] = [round(x/msn02.shape[0]*100, 2) for x in list(msn02.isnull().sum())]
data1['n_unique'] = list(msn02.nunique())

df = pd.DataFrame(data1)
df
```

```python
# Convert 'UTC_TIME' to datetime
msn02['UTC_TIME'] = pd.to_datetime(msn02['UTC_TIME'])

# Create new columns for date and time
msn02['DATE'] = msn02['UTC_TIME'].dt.date
msn02['TIME'] = msn02['UTC_TIME'].dt.time

# Display the DataFrame to verify the changes
print(msn02[['UTC_TIME', 'DATE', 'TIME']].head())
```

```python
nan_count = msn02['FLIGHT_PHASE_COUNT'].isnull().sum()
print("Number of NaN values in 'FLIGHT_PHASE_COUNT':", nan_count)
```

```python
flight_phase_counts = msn02['FLIGHT_PHASE_COUNT'].value_counts()
print(flight_phase_counts)
```

```python
# Dropping rows where 'FLIGHT_PHASE_COUNT' is NaN
msn02 = msn02.dropna(subset=['FLIGHT_PHASE_COUNT'])

# Display the DataFrame to verify the changes
msn02.head()
```

```python
# Check for duplicates in 'UTC_TIME'
duplicate_count = msn02['UTC_TIME'].duplicated().sum()
print("Number of duplicate values in 'UTC_TIME':", duplicate_count)
```

```python
# Dropping duplicate rows based on 'UTC_TIME'
msn02 = msn02.drop_duplicates(subset='UTC_TIME', keep='first')

# Display the DataFrame to verify the changes
msn02.head()
```

```python
# Filter the DataFrame to include only rows where 'FLIGHT_PHASE_COUNT' is 8
phase_8_df = msn02[msn02['FLIGHT_PHASE_COUNT'] == 8]

# Display the DataFrame
phase_8_df.head(10)
```

```python

pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None)

# Creating the summary DataFrame for phase_8_df
data_phase_8 = {}
data_phase_8['Features'] = list(phase_8_df.columns)
data_phase_8['phase_8_missing'] = list(phase_8_df.isnull().sum())
data_phase_8['phase_8_missing_%'] = [round(x/phase_8_df.shape[0]*100, 2) for x in list(phase_8_df.isnull().sum())]
data_phase_8['n_unique'] = list(phase_8_df.nunique())

summary_phase_8_df = pd.DataFrame(data_phase_8)

# Display the summary DataFrame
summary_phase_8_df
```

```python
# Adding the FUEL_FLOW columns and creating FUEL_FLOW_ADDED
phase_8_df['FUEL_FLOW_ADDED'] = phase_8_df[['FUEL_FLOW_1', 'FUEL_FLOW_2', 'FUEL_FLOW_3', 'FUEL_FLOW_4']].sum(axis=1)

# Adding the FUEL_USED columns and creating FUEL_USED_ADDED
phase_8_df['FUEL_USED_ADDED'] = phase_8_df[['FUEL_USED_1', 'FUEL_USED_2', 'FUEL_USED_3', 'FUEL_USED_4']].sum(axis=1)

# Display the DataFrame to verify the changes
phase_8_df.head(10)
```

```python
# List of columns to be summed up
fuel_qty_columns = [
                    'VALUE_FUEL_QTY_CT', 'VALUE_FUEL_QTY_FT1', 'VALUE_FUEL_QTY_FT2', 
                    'VALUE_FUEL_QTY_FT3', 'VALUE_FUEL_QTY_FT4', 'VALUE_FUEL_QTY_LXT', 'VALUE_FUEL_QTY_RXT']

# Adding the specified columns and creating VALUE_FUEL_QTY_ADDED
phase_8_df['VALUE_FUEL_QTY_ADDED'] = phase_8_df[fuel_qty_columns].sum(axis=1)

# Display the DataFrame to verify the changes
phase_8_df.head(10)
```

```python
phase_8_df['Flight'].value_counts()
```

```python
# Filter the DataFrame to include only rows where 'Flight' is 'V0965'
flight_v0965_df = phase_8_df[phase_8_df['Flight'] == 'V0965']

# Display the DataFrame
flight_v0965_df.head()
```

```python
# Selecting specific columns to create the new DataFrame
data_df = flight_v0965_df[['UTC_TIME', 'Flight', 'DATE', 'TIME', 
                           'FUEL_FLOW_ADDED', 'FUEL_USED_ADDED', 
                           'VALUE_FUEL_QTY_ADDED', 'VALUE_FOB']]

# Display the DataFrame to verify the changes
data_df.head(10)
```

```python
# Subtracting VALUE_FUEL_QTY_ADDED from VALUE_FOB and creating the new column
data_df['DIFF_FOB_VALUE_FUEL'] =  data_df['VALUE_FUEL_QTY_ADDED'] - data_df['VALUE_FOB']

# Display the DataFrame to verify the changes
data_df.head()
```

```python
data_df.head()
```

```python
data_df['FUEL_USED_FOB'] = 0
```

```python
data_df.head()
```

```python

# Calculate the difference between the previous row and the current row in VALUE_FOB
data_df['FUEL_USED_FOB'] = data_df['VALUE_FOB'].shift(1) - data_df['VALUE_FOB']

# Set the first row of FUEL_USED_FOB to 0
data_df['FUEL_USED_FOB'].iloc[0] = 0

```

```python
data_df.head()
```

```python
# Calculate the difference between the previous row and the current row in VALUE_FOB
data_df['FUEL_USED'] = data_df['FUEL_USED_ADDED'].shift(-1) - data_df['FUEL_USED_ADDED']

# Set the first row of FUEL_USED_FOB to 0
data_df['FUEL_USED'].iloc[0] = 0
```

```python
data_df.head()
```

```python
data_df['leakage']= data_df['FUEL_USED_FOB'] - data_df['FUEL_USED']

data_df.head()
```

```python
# Save the grouped DataFrame to a CSV file
data_df.to_csv('data_df_V095.csv', index=False)
```
